```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
```

```python
iris = datasets.load_iris()
x = iris.data[:,:2]
y = iris.target

x = x[y!=2]
y = y[y!=2]
model = SVC(kernel='linear')
model.fit(x,y)
```

```python
w = model.coef_[0]
b = model.intercept_[0]
print("Weights:", w)
print("Bias:",b)
print("Supoort Vectors:", model.support_vectors_)
plt.scatter(x[:,0], x[:,1], c = y, cmap = "coolwarm")
```

```python
plt.scatter(
    model.support_vectors_[:,0],
    model.support_vectors_[:,1],
    s = 100,
    facecolors = "none",
    edgecolors = "k",
    label = "support vectors"
)

x0 = np.linspace(x[:,0].min(), x[:,0].max(), 100)
x1 = -(w[0]*x0 + b)/w[1]

x1_p = -(w[0]*x0 + b - 1)/w[1]
x1_n = -(w[0]*x0 + b + 1)/w[1]

plt.plot(x0, x1, "k", label = "Decision Boundary")
plt.plot(x0, x1_p,"r--", label = "Positve Margin")
plt.plot(x0, x1_n, "b--", label = "Negative Margin" )

plt.legend()
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.title("SVM")
plt.plot(x0,x1)
plt.show()
```
