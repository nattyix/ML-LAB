```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
```

```python
centers = [[34,-118], [40,-74], [41,-87]]
x,_ = make_blobs(n_samples = 300, centers = centers, cluster_std = 0.5, random_state = 42)
df = pd.DataFrame(x, columns = ['Latitude', 'Longitude'])
dbscan = DBSCAN(eps = 0.3, min_samples = 5)

df['cluster'] = dbscan.fit_predict(df)
```

```python
plt.scatter(df['Longitude'], df['Latitude'], c = df['cluster'])
plt.title("DBSCAN")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
clusters = len(set(df['cluster'])) - (1 if -1 in df['cluster'].values else 0)
noise = list(df['cluster']).count(-1)
print("Clusters:", clusters)
print("Noise :", noise)
```
